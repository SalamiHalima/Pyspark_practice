```
# bring everything together, wanna create a new col with conditional statement

from pyspark.sql.functions import when
```

```
data = data.withColumn('Profitable_flag', 
    when(data['Profit'] >= 0, 'Profitable').
    when(data['Profit'] < 0, 'Non Profitable').
    when(data['Profit'] == 0, 'Zero').
    otherwise('Error'))
```

```
grouped_country = data.groupBy('Country').agg({'Sales': 'sum'}).orderBy('sum(Sales)')
grouped_country.show()
```

```
# collect returns -> all rows from all different executors into the memory and it returns them in a list form so we can iterate.
#Note: You can only use the .collect() function when the data is completely grouped into shorter form because it will take so long if not.

grouped_country.collect()  #[:3]
```

```
bottom_country_sales = grouped_country.collect()[0][0]
bottom_country_sales
```

```
# show me all the row of the country which has the lowest sales
```

```
data.filter(data['Country'] == bottom_country_sales).show()
```

```
# show me all the row of the bottom 3 countries which has the lowest profit
```

```
bottom3_country_profit = data.groupBy('Country').agg({'Profit': 'sum'}).orderBy('sum(Profit)').collect()[:3]
bottom3_country_profit
```

```
# iterate through to get the all the rows of the country using a list comprehension

bottom_countries = [every[0] for every in bottom3_country_profit]
bottom_countries
```

```
data.filter(data['Country'].isin(bottom_countries)).show()
```

```
# How to change the format of a date in pyspark

from pyspark.sql import functions as f
```

```
# we want to make a new col for it as we changing the date format

data.withColumn('clean order date', f.to_date(data['Order Date'], 'dd/MM/yyyy')).select('Order Date', 'clean order date').printSchema()
```

```
# backlash \ at the end to continue in the next line helps well without given an err.

data = data.withColumn('clean order date', f.to_date(data['Order Date'], 'dd/MM/yyyy')).\
       withColumn('clean ship date', f.to_date(data['Ship Date'], 'dd/MM/yyyy'))
#data.show()
```

```
# You can now order the clean order date.

data.select('clean ship date', 'clean order date').orderBy('clean order date').printSchema()#show(5)
```

```
# udf -> user Define Function 
# Note: udf is an expensive function in pyspark and it is preferrable to apply it on grouped data.

from pyspark.sql.functions import udf

# which makes you define the function you want
```

```
# we want to create a new col for customer name with uppercase all through

def makeupper(mystring):
    return mystring.upper()

upperUDF = udf(lambda x : makeupper(x))

# create a new col for the uppercase of the col you wanna use

data = data.withColumn('upper case name', upperUDF(data['Customer Name']))
data.show()
```

```
data.select('Customer Name', 'upper case name').show(5)
```
